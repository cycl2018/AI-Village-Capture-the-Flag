# Math Challenge 4 (id: math_4) -- 400 Points

What's the dimensionality of the clusters in clusters1.npy? The key is the dimensions ordered by cluster size (smallest to largest), with no spaces.  These files are available in the input directory under `math`.

```python
import itertools
list(itertools.permutations(range(1,3)))
```

```python
flag = 0
for i in range(1,10):
    ans = list(itertools.permutations(range(1,i+1)))
    for res in ans:
        dim = ''.join(list(map(str,res)))
        try:
            r = requests.post("http://flags.fly.dev:5000/score", json=json.dumps({"challenge_id": "math_4", "submission": f"{dim}"}))
            print(dim,r.text)
            if len(r.text) > 100:
                flag = 1
                break
        except requests.exceptions.ConnectionError:
            print("Connection problems. Contact the CTF organizers.")
    if flag > 0:
        break 
```
