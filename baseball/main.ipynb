```python
mp = {'pitches': []}
for i in range(15):
    mp['pitches'].append({"x":random.randint(0, 29), "y":random.randint(0, 29)})
```

```python
def gt_num(s):
    if '0.' not in s:
        return -10000.0
    res = '0.'
    f = 0
    for i in s:
        if f == 1:
            f = 2
            continue
        if i == '0' and f == 0:
            f = 1
            continue
        if f == 2:
            if ord('0') <= ord(i) <= ord('9'):
                res += i
    return float(res)
```

```python
best_pro = 100.0
f = 0
for _ in range(3000):
    i = random.randint(0, 14)
    x = random.randint(0, 29)
    y = random.randint(0, 29)
    tx = mp['pitches'][i]["x"]
    ty = mp['pitches'][i]["y"]
    mp['pitches'][i]["x"] = x
    mp['pitches'][i]["y"] = y

    json.dump(mp, open("./pitch.json", 'w+'))
    test = json.load(open("./pitch.json", 'r+'))
    r = requests.post("https://bball.fly.dev/score", json=test)
    
    res = json.loads(r.text)
    pro = gt_num(res['message'][0])
    
    print(res, f, pro, best_pro)

    if pro < 0:
        mp['pitches'][i]["x"] = tx
        mp['pitches'][i]["y"] = ty
        continue
    
    if 'enry' in res['message'][0]:
        if f == 0:
            f = 1
            best_pro = pro
            continue
        if pro > best_pro:
            best_pro = pro
        else:
            mp['pitches'][i]["x"] = tx
            mp['pitches'][i]["y"] = ty           
    else:
        if f == 1:
            mp['pitches'][i]["x"] = tx
            mp['pitches'][i]["y"] = ty  
            continue
        if pro < best_pro:
            best_pro = pro
        else:
            mp['pitches'][i]["x"] = tx
            mp['pitches'][i]["y"] = ty    
```
